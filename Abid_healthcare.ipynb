```python
import pandas as pd
from rdflib import Namespace, Graph, URIRef, RDF, RDFS, Literal
```

```python
filepath = "Diagnostic_Test_Results_abid.csv"
ontology_file = "MappOntology_abid.ttl"
df = pd.read_csv(filepath, delimiter=';')
df.head()
```

```python
# #Reference variables
ns=Namespace('http://www.semanticweb.org/abid/ontologies/2023/7/CardiovascularDiseaseOntology#')
rdf = Namespace('http://www.w3.org/1999/02/22-rdf-syntax-ns#')
rdfs = Namespace('http://www.w3.org/2000/01/rdf-schema#')

#Variables

new_id = URIRef(ns+'Id')
new_age= URIRef(ns+'Age')
new_gender = URIRef(ns+'Gender')
new_Pheight = URIRef(ns+'Patient_Height')
new_Pweight = URIRef(ns+'Patient_Weight')
new_Systolic_BP = URIRef(ns+'Systolic_BP')
new_Diastolic_BP = URIRef(ns+'Diastolic_BP')
new_choles = URIRef(ns+'Cholestrol')
new_gluc = URIRef(ns+'Glucose')
new_smoke = URIRef(ns+'Smoke')
new_alco = URIRef(ns+'Alcohol')
new_active = URIRef(ns+'Active')
new_cardio = URIRef(ns+'Cardio')
diagnosticClass = URIRef(ns+'Diagnostic_Test_Results')

ontologyTriples = []

for index, row in df.iterrows():
  #  print (row.age)
    Id = Literal(int(row['Id']))
    test = URIRef(ns+'Diagnostic_Test_Results_'+ str(row['Id']))
    ontologyTriples.append((test, RDF.type, diagnosticClass)) 
    
    age = Literal(int(row['Age']), )
    ontologyTriples.append((test, new_age, age))
      
    gender = Literal(int(row['Gender']))
    ontologyTriples.append((test, new_age, gender))
    
    Pheight = Literal(int(row['Patient_Height']))
    ontologyTriples.append((test, new_Pheight, Pheight))
    
    Pweight = Literal(int(row['Patient_Weight']))
    ontologyTriples.append((test, new_Pweight, Pweight))
    
    Systolic_BP= Literal(int(row['Systolic_BP']))
    ontologyTriples.append((test, new_Systolic_BP, Systolic_BP))
     
    Diastolic_BP = Literal(int(row['Diastolic_BP']))
    ontologyTriples.append((test, new_Diastolic_BP, Diastolic_BP))
    
    choles = Literal(int(row['Cholesterol']))
    ontologyTriples.append((test, new_choles, choles))
   
    gluc = Literal(int(row['Glucose']))
    ontologyTriples.append((test, new_gluc, gluc))
         
    smoke = Literal(int(row['Smoke']))
    ontologyTriples.append((test, new_smoke, smoke))
    
    alco = Literal(int(row['Alcohol']))
    ontologyTriples.append((test, new_alco, alco))
    
    active = Literal(int(row['Active']))
    ontologyTriples.append((test, new_active, active))
    
    cardio = Literal(int(row['Cardio']))
    ontologyTriples.append((test, new_cardio, cardio))
```

```python
ontologyTriples
```

```python
def extend_owl_turtle(owl_file, triples_to_add):
    # Create an RDF graph
    graph = Graph()

    # Load the existing OWL Turtle file
    print(owl_file, graph)
    try:
        graph.parse(owl_file, format="turtle")
    except FileNotFoundError:
        print(f"Error: {owl_file} not found.")
        return

    # Add new triples to the graph
    for triple in triples_to_add:
        subject, predicate, obj = triple
        graph.add((subject, predicate, obj))

    # Save the updated graph to the OWL Turtle file
    graph.serialize(destination=owl_file, format="turtle")
    print('Ontology population has been done successfully')
```

```python
extend_owl_turtle(ontology_file, ontologyTriples)
```
